# Aggregation

* Calculate Mean, Median, Min, Max for cpc column

```python
data.groupby('Country')['cpc'].agg([np.min, np.max, np.mean, np.median])
```

* Calculate Mean, Median, Min, Max for cpc and pop

```python
data.groupby('Country')[['cpc', 'pop']].agg([np.min, np.max, np.mean, np.median])
```
