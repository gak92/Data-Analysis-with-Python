##Task 3:
* Data Aggregation using GroupBy. 
* Using the datafile "Kiwisaver_sample_wk9.csv", group the Funds based on Branding Name and Equity Style. 
* Report the average Holdings and Net Assets Size of the groups that you just created.

```python
file_path = '/content/drive/MyDrive/AI/Data Analysis/DA with Python/Kiwisaver_sample_wk9.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
df_branding_name_equity_style = data.groupby(['Branding Name', 'Equity Style'])[['Holdings', 'Net Assets Size']].mean()
df_branding_name_equity_style
```

```python
df_branding_name = data.groupby(['Branding Name'])[['Holdings', 'Net Assets Size']].mean()
df_branding_name
```

```python
df_equity_style = data.groupby(['Equity Style'])[['Holdings', 'Net Assets Size']].mean()
df_equity_style
```

##Task 5:
* Still using the file "Kiwisaver_sample_wk9.csv", clean the dataframe if information related to Morningstar Rating and Manager Name is not available.

* Using either GroupBy or a pivot table, report the average Morningstar Rating for Funds managed by each of the managers in the file. 
* Plot these average Morningstar Rating as a barplot

```python
data.head()
```

```python
data_cleaned = data[['Morningstar Rating', 'Manager Name']].dropna()
data_cleaned.head()
```

```python
data_cleaned_group = data_cleaned.groupby('Manager Name')['Morningstar Rating'].mean()
data_cleaned_group
```

```python
data_cleaned_group.plot(kind='bar')
plt.show()
```

```python
data_cleaned_pivot = data_cleaned.pivot_table(values=['Morningstar Rating'], index=['Manager Name'])
data_cleaned_pivot
```

```python
data_cleaned_pivot.plot(kind='bar')
plt.show()
```
