# Using Beautifulsoup

```python
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.python.org/~guido/'
```

```python
response = requests.get(URL)
print(response)
```

## HTML RAW

```python
html_doc = response.text
print(html_doc)
```

## Beautifulsoup

```python
bs = BeautifulSoup(html_doc)
print(bs)
```

## Prettify

```python
bs_prettify = bs.prettify()
print(bs_prettify)
```

## Get the Title

```python
title = bs.title
print(title)
```

## Get the Text

```python
text = bs.get_text()
print(text)
```

## Find all a tags

```python
a_tags = bs.find_all('a')
print(a_tags)
```

```python
for link in a_tags:
  print(link.get('href'))
```
