# Reading HDF5 File

```python
import h5py
```

```python
hdf5_file = '/content/drive/MyDrive/AI/Data Analysis/DA with Python/LIGO_data.hdf5'
```

```python
data_hdf5 = h5py.File(hdf5_file, mode='r')
print(type(data_hdf5))
data_hdf5
```

```python
for key in data_hdf5.keys():
  print(key)
```

```python
for k,v in data_hdf5.items():
  print(k, v)
```

## Reading groups

```python
with h5py.File(hdf5_file, mode='r') as hdf:
  items = list(hdf.items())
  print("ITEMS: \n", items)

  # Getting a group
  meta = hdf.get('meta')
  meta_items = list(meta.items())
  print('\nMETA:\n', meta_items)

  # Reading dataset in group meta
  ds1 = meta.get('Description')
  ds1 = np.array(ds1)
  print('\nDATASET:\n', ds1)
```

## Reading other group (strain)

```python
group = data_hdf5['strain']
print("GROUP: ",group)

for key in group.keys():
  print("KEY: ", key)

print("DATASET: ", data_hdf5['strain']['Strain'])
ds_strain = data_hdf5['strain']['Strain'][:]
print("ds_strain Type: ", type(ds_strain))
print("Length of Dataset: ", len(ds_strain))
ds_strain[:10]
```

# Plot Strain Time Series Data

```python
num_samples = 10000
time = np.arange(0, 1, 1/num_samples)

plt.plot(time, ds_strain[:num_samples])
plt.xlabel('GPS Time')
plt.ylabel('Strain Data')
plt.show()
```

# Reading Database File

```python
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

```python
sql_file = '/content/drive/MyDrive/AI/Data Analysis/DA with Python/Chinook.sqlite'
```

```python
engine = create_engine('sqlite:///'+sql_file)
```

```python
table_names = engine.table_names()
table_names
```

```python
insp = inspect(engine)
table_names = insp.get_table_names()
print(table_names)
```

## Make a connection & execute query

```python
conn = engine.connect()
query = conn.execute('SELECT * FROM Album')

df = pd.DataFrame(query.fetchall())
conn.close()

df
```

## Fetching few records

```python
with engine.connect() as conn:
  query = conn.execute('SELECT LastName, Title FROM Employee')
  df = pd.DataFrame(query.fetchmany(size=3))
  df.columns = query.keys()

df
```

## Fetching records with condition

```python
with engine.connect() as conn:
  query = conn.execute('SELECT * FROM Employee WHERE EmployeeID >= 6')
  df = pd.DataFrame(query.fetchall())
  df.columns = query.keys()

df
```

## Order By

```python
with engine.connect() as conn:
  query = conn.execute('SELECT * FROM Employee ORDER BY BirthDate')
  df = pd.DataFrame(query.fetchall())
  df.columns = query.keys()

df
```

## Reading using pandas

```python
df1 = pd.read_sql_query('SELECT * FROM Album', engine)
df1.head()
```

```python
with engine.connect() as conn:
  query = conn.execute('SELECT *FROM Album')
  df2 = pd.DataFrame(query.fetchall())
  df2.columns = query.keys()

df2.head()
```

```python
df1.equals(df2)
```

## Complex query

```python
df = pd.read_sql_query('SELECT * FROM Employee WHERE EmployeeId>=6 ORDER BY BirthDate', engine)

df.head()
```

## INNER JOIN

```python
with engine.connect() as conn:
   query = conn.execute("SELECT Title, Name FROM Album INNER JOIN Artist on Album.ArtistID = Artist.ArtistID")
   df = pd.DataFrame(query.fetchall())
   df.columns = query.keys()

df.head()
```

```python
df = pd.read_sql_query('SELECT * FROM PlaylistTrack INNER JOIN Track on PlaylistTrack.TrackId = Track.TrackId WHERE Milliseconds < 250000', engine)
df.head()
```
