# Appending

```python
data1 = pd.read_csv(data1_path)
data2 = pd.read_csv(data2_path)
```

```python
data1
```

```python
data2
```

```python
cpc_data = data1['cpc']
cpc_data
```

```python
gold_data = data2['gold']
gold_data
```

```python
quarter1 = cpc_data.append(gold_data)
quarter1
```

```python
print(quarter1.sum())
```

# Build List of Series

```python
units = []
for i in [cpc_data, gold_data]:
  units.append(i)

print(units)
```

# Concatenate List

```python
quarter1 = pd.concat(units, axis='rows')
quarter1
```

```python
quarter2 = pd.concat(units, axis='columns')
quarter2
```
